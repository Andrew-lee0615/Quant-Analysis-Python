```python
import pandas as pd
import numpy as np

# 1. 샘플 데이터 생성
data = {
    'Date': pd.date_range(start='2026-01-01', periods=10, freq='D'),
    'Price': [420, 425, 430, 415, 410, 405, 420, 415, 400, 405]
}
df = pd.DataFrame(data)

# 2. MDD 계산 로직
df['Peak'] = df['Price'].cummax()               # 누적 고점 계산
df['Drawdown'] = (df['Price'] / df['Peak']) - 1  # 고점 대비 하락률
mdd = df['Drawdown'].min()                      # 최저점이 곧 MDD

print(f"MSFT 분석 결과:")
print(f"최대 낙폭(MDD): {mdd:.2%}")
df[['Date', 'Price', 'Peak', 'Drawdown']]
```
