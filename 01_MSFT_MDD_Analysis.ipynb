```python
import pandas as pd
import numpy as np

# 1. 샘플 데이터 생성
data = {
    'Date': pd.date_range(start='2026-01-01', periods=10, freq='D'),
    'Price': [420, 425, 430, 415, 410, 405, 420, 415, 400, 405]
}
df = pd.DataFrame(data)

# 2. MDD 계산 로직
df['Peak'] = df['Price'].cummax()               # 누적 고점 계산
df['Drawdown'] = (df['Price'] / df['Peak']) - 1  # 고점 대비 하락률
mdd = df['Drawdown'].min()                      # 최저점이 곧 MDD

print(f"MSFT 분석 결과:")
print(f"최대 낙폭(MDD): {mdd:.2%}")
df[['Date', 'Price', 'Peak', 'Drawdown']]
```

```python
import matplotlib.pyplot as plt

# 1. 차트 그리기 (시각화 라이브러리 matplotlib 사용)
plt.figure(figsize=(10, 5))

# 주가 그래프
plt.subplot(2, 1, 1)
plt.plot(df['Date'], df['Price'], marker='o', label='MSFT Price')
plt.plot(df['Date'], df['Peak'], linestyle='--', color='red', label='Historical Peak')
plt.title('MSFT Price & Peak')
plt.legend()

# 낙폭(Drawdown) 그래프
plt.subplot(2, 1, 2)
plt.fill_between(df['Date'], df['Drawdown'], 0, color='red', alpha=0.3)
plt.title('Drawdown (%)')

plt.tight_layout()
plt.show()
```
